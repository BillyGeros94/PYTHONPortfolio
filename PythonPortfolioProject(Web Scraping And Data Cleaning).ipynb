```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_temp = pd.read_html('https://en.wikipedia.org/wiki/Forbes_list_of_the_most_valuable_NBA_clubs')
```

```python
df1 = df_temp[1]
```

```python
df1
```

```python
df1.drop({'Rank', 'Swing[a]', 'Change[c]', 'Debt[d]'}, inplace = True, axis = 1)
```

```python
df1.rename(columns={'Value[b]': 'Value', 'Revenue[e]': 'Revenue', 'Operating Income[f]': 'Operating Income'}, inplace=True)
```

```python
df1['Sport'] = 'Basketball'
df1['League'] = 'NBA'
df1.insert(loc=1, column='Country', value='United States')
```

```python
df1['Value'] = df1['Value'].str.replace('4.3 billion', '$4.3 billion')
```

```python
df1
```

```python
df_temp = pd.read_html('https://en.wikipedia.org/wiki/Forbes_list_of_the_most_valuable_football_clubs')
```

```python
df2 = df_temp[0]
```

```python
df2
```

```python
df2.drop({'Rank 2023','Rank 2024', '% change on year', 'Debt as % of value' }, inplace = True, axis = 1)
```

```python
df2.rename(columns={'Value in $ millions': 'Value', 'Revenue ($M)': 'Revenue', 'Operating income ($M)': 'Operating Income'}, inplace=True)
```

```python
df2['Sport'] = 'Association football'

df2['League'] = ('La Liga', 'Premier League', 'La Liga', 'Premier League', 'Premier League', 'Bundesliga', ' Ligue 1', 
                 'Premier League', 'Premier League', 'Premier League', 'Serie A', ' Bundesliga', 'La Liga', 'Serie A',
                 'MLS', 'Premier League', 'MLS', 'Serie A', 'MLS', 'MLS')

df2.insert(loc=2, column='Territory', value=['Madrid', 'Manchester', 'Barcelona', 'Liverpool', 'Manchester', 'Munich', 'Paris', 'London', 'London', 
                                            'London', 'Turin', 'Dortmund', 'Madrid', 'Milan', 'California', 'London', 'Florida', 'Milan', 'California', 
                                            'Georgia'])
```

```python
df2['Value'] = df2['Value']/1000
```

```python
df2['Value'] = '$' + df2['Value'].astype(str) + ' billion'
df2['Revenue'] = '$' + df2['Revenue'].astype(str) + ' million'
df2['Operating Income'] = '$' + df2['Operating Income'].astype(str) + ' million'
```

```python
df2['Operating Income'] = df2['Operating Income'].str.replace('$-', '-$')
```

```python
df2
```
