```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\Vasilis\Downloads\laptop_prices.csv')
```

```python
df
```

```python
df.info()
```

```python
df.describe().drop(['laptop_ID'], axis = 1)
```

```python
df.groupby('Company').describe().drop(['laptop_ID'], axis = 1)
```

```python
df.groupby('Company')['Price_usd'].mean().sort_values(ascending = False)
```

```python
df1 = df.groupby('Company', as_index = False).agg({'Price_usd':'mean'}).sort_values(by = 'Price_usd')

plt.figure(figsize = (20,10))
sns.set_style('whitegrid')
sns.lineplot( x = 'Company', y = 'Price_usd', data = df1, marker = 'o', markersize=10, color = '#022323')
plt.title('Average Laptop Price Per Company', fontsize=14, fontweight='bold')
plt.xlabel('Company')
plt.ylabel('Price (USD)')
```

```python
df.groupby('TypeName')['Price_usd'].mean().sort_values()
```

```python
df2 = df.groupby('TypeName', as_index = False).agg({'Price_usd':'mean'}).sort_values(by = 'Price_usd')

plt.figure(figsize = (20,10))
sns.set_style('darkgrid')
sns.barplot( x = 'TypeName', y = 'Price_usd', data = df2, color = '#ff022f')
plt.title('Average Laptop Price Per Type', fontsize=14, fontweight='bold')
plt.xlabel('Laptop Type')
plt.ylabel('Price (USD)')
```

```python
df.groupby('Ram')['Price_usd'].mean().sort_values()
```

```python
df3 = df.groupby('Ram', as_index = False).agg({'Price_usd':'mean'})
df3['Ram'] = df3['Ram'].str.replace('GB', '').astype(int)

plt.figure(figsize = (20,10))
sns.set_style('white')
sns.pointplot( x = 'Ram', y = 'Price_usd', data = df3,color = '#5e6667')
plt.title('Average Laptop Price Per Ram Capacity', fontsize=14, fontweight='bold')
plt.xlabel('Ram (GB)')
plt.ylabel('Price (USD)')
```

```python
df.groupby('OpSys')['Price_usd'].mean().sort_values()
```

```python
df4 = df.groupby('OpSys', as_index = False).agg({'Price_usd':'mean'}).sort_values(by = 'Price_usd')

plt.figure(figsize = (20,10))
sns.set_style('dark')
sns.barplot( x = 'Price_usd', y = 'OpSys', data = df4, color = '#156062')
plt.title('Average Laptop Price Per Operating System', fontsize=14, fontweight='bold')
plt.xlabel('Price (USD)')
plt.ylabel('Operating System')
```
