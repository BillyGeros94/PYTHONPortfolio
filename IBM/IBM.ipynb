```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
balance_sheet = pd.read_csv(r'C:\Users\Vasilis\Downloads\IBM\balance_sheet_IBM.csv')
cash_flow = pd.read_csv(r'C:\Users\Vasilis\Downloads\IBM\cash_flow_data.csv')
income_statement = pd.read_csv(r'C:\Users\Vasilis\Downloads\IBM\income_statement_IBM.csv')
```

```python
balance_sheet['fiscalDateEnding'] = pd.to_datetime(balance_sheet['fiscalDateEnding'], dayfirst=True)
cash_flow['fiscalDateEnding'] = pd.to_datetime(cash_flow['fiscalDateEnding'])
income_statement['fiscalDateEnding'] = pd.to_datetime(income_statement['fiscalDateEnding'], dayfirst=True)
```

```python
balance_sheet.fillna(0, inplace = True)

cash_flow.fillna(0, inplace = True)

income_statement.fillna(0, inplace = True)
```

```python
balance_sheet['current_ratio'] = balance_sheet['totalCurrentAssets'] / balance_sheet['totalCurrentLiabilities']
balance_sheet['quick_ratio'] = (balance_sheet['totalCurrentAssets'] - balance_sheet['inventory']) / balance_sheet['totalCurrentLiabilities']

balance_sheet[['fiscalDateEnding', 'current_ratio', 'quick_ratio']]
```

```python
income_statement['gros_profit_margin'] = (income_statement['totalRevenue'] - income_statement['costofGoodsAndServicesSold']) / income_statement['totalRevenue']

income_statement['gros_profit_margin']
```

```python
balance_sheet['assets_growth'] = balance_sheet['totalAssets'].pct_change() * 100
balance_sheet['liabilities_growth'] = balance_sheet['totalLiabilities'].pct_change() * 100
balance_sheet['equity_growth'] = balance_sheet['totalShareholderEquity'].pct_change() * 100


balance_sheet[['fiscalDateEnding', 'assets_growth', 'liabilities_growth', 'equity_growth']]
```

```python
balance_sheet['cash_ratio'] = balance_sheet['cashAndCashEquivalentsAtCarryingValue'] / balance_sheet['totalCurrentLiabilities']

balance_sheet['cash_ratio']
```

```python
income_statement['gross_profit_margin'] = (income_statement['grossProfit'] / income_statement['totalRevenue']) * 100

income_statement['net_profit_margin'] = (income_statement['netIncome'] / income_statement['totalRevenue']) * 100

income_statement[['gross_profit_margin', 'net_profit_margin']]
```

```python
balance_sheet['debt_to_equity'] = balance_sheet['totalLiabilities'] / balance_sheet['totalShareholderEquity']

balance_sheet['debt_to_equity']
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='fiscalDateEnding', y='totalAssets', data=balance_sheet, label='Total Assets')
sns.lineplot(x='fiscalDateEnding', y='totalLiabilities', data=balance_sheet, label='Total Liabilities')
sns.lineplot(x='fiscalDateEnding', y='totalShareholderEquity', data=balance_sheet, label='Shareholder Equity')
plt.title('IBM Balance Sheet Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Value (in millions)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='fiscalDateEnding', y='operatingCashflow', data=cash_flow, label='Operating Cash Flow')
sns.lineplot(x='fiscalDateEnding', y='cashflowFromInvestment', data=cash_flow, label='Investment Cash Flow')
sns.lineplot(x='fiscalDateEnding', y='cashflowFromFinancing', data=cash_flow, label='Financing Cash Flow')
plt.title('IBM Cash Flow Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Cash Flow (in millions)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.stackplot(balance_sheet['fiscalDateEnding'], balance_sheet['totalAssets'], balance_sheet['totalLiabilities'], balance_sheet['totalShareholderEquity'],
              labels=['Total Assets', 'Total Liabilities', 'Equity'], alpha=0.6)

plt.title('IBM Capital Structure Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Value (in millions)')
plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='fiscalDateEnding', y='operatingCashflow', data=cash_flow, label='Operating Cash Flow')
sns.lineplot(x='fiscalDateEnding', y='cashflowFromInvestment', data=cash_flow, label='Investment Cash Flow')
sns.lineplot(x='fiscalDateEnding', y='cashflowFromFinancing', data=cash_flow, label='Financing Cash Flow')

plt.title('IBM Cash Flow Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Cash Flow (in millions)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='fiscalDateEnding', y='current_ratio', data=balance_sheet, label='Current Ratio')
sns.lineplot(x='fiscalDateEnding', y='quick_ratio', data=balance_sheet, label='Quick Ratio')
sns.lineplot(x='fiscalDateEnding', y='cash_ratio', data=balance_sheet, label='Cash Ratio')

plt.title('IBM Liquidity Ratios Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Ratio')
plt.legend()
plt.show()
```

```python
cash_flow['free_cash_flow'] = cash_flow['operatingCashflow'] - cash_flow['capitalExpenditures']

plt.figure(figsize=(10,6))
sns.lineplot(x='fiscalDateEnding', y='free_cash_flow', data=cash_flow, label='Free Cash Flow')

plt.title('IBM Free Cash Flow Over Time')
plt.xlabel('Fiscal Year')
plt.ylabel('Free Cash Flow (in millions)')
plt.legend()
plt.show()
```

```python
corr_matrix = balance_sheet.drop('reportedCurrency', axis = 1).corr()

mask = np.triu(np.ones_like(corr_matrix, dtype = bool))

plt.figure(figsize=(30, 30))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', mask = mask, square = True, linewidths = 0.5)
plt.title('Correlation Heatmap - IBM Balance Sheet')
plt.show()
```
