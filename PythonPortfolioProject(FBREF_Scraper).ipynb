```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import csv
```

```python
URL = 'https://fbref.com/en/comps/9/2023-2024/2023-2024-Premier-League-Stats'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
soup = BeautifulSoup(soup.prettify(), 'html.parser' )
```

```python
table= soup.find_all('table')[2]
```

```python
table_titles = table.find_all('th')
table_titles = [tittle.text.strip() for tittle in table_titles]
```

```python
df = pd.DataFrame( columns = table_titles[7:38])
index = table_titles[38:]
```

```python
column_data = table.find_all('tr')[2:22]

for row in column_data:
    row_data = row.find_all('td')
    table_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = table_data
```

```python
df.insert(0, 'Squad', index)
```
